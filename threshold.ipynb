```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('test4.png')
```

```python
def show_pic(img):
        fig = plt.figure(figsize=(10,10))
        ax = fig.add_subplot(111)
        ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python

def display_img(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Load an example image

# Define a Laplacian kernel for edge detection
kernel = np.array([[1, 1, 1],
                   [1, -8, 1],
                   [1, 1, 1]], dtype=np.float32)

# Apply the convolution filter
dst = cv2.filter2D(img, -1, kernel)

# Display the result
display_img(dst)
```

```python


def display_img(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Load an example image
# Convert image to float32 for accurate multiplication
img_float = img.astype(np.float32)

# Increase brightness by 120%
brightness_factor = 0.5
brighter_img = img_float * brightness_factor

# Clip values to stay within valid range [0, 255]
brighter_img = np.clip(brighter_img, 0, 255)

# Convert image back to uint8
brighter_img = brighter_img.astype(np.uint8)

# Display the result
display_img(brighter_img)
```
